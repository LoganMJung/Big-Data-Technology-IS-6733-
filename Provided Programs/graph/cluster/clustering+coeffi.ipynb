```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
G4Clique = nx.complete_graph(6)
nx.draw_networkx(G4Clique,node_color='cyan')
print(nx.clustering(G4Clique))
```

```python
star_graph = nx.star_graph(6)
nx.draw_networkx(star_graph,node_color='cyan')
print(nx.clustering(star_graph))
```

```python
G1 = nx.Graph()
G1.add_edges_from([(0, 1),
                   (2, 1),
                   (2, 6),
                   (2, 3),
                   (0, 6),
                   (3, 6),
                   (0, 2),
                   (0, 3),
                   (0, 5),
                   (1, 3),
                   (1, 6),
                   (3, 4),
                   (4, 5),
                   (4, 7),
                   (5, 8),
                   (8, 9)])
nx.draw_networkx(G1,node_color='cyan')
print(nx.clustering(G1))
```
