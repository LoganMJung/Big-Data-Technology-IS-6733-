```python
import networkx as nx
%matplotlib notebook
```

## Undirected Graph Network

```python
graph=nx.Graph()
graph.add_edge('A','B')
graph.add_edge('B','C')
graph.add_edge('C','F')
graph.add_edge('C','E')
graph.add_edge('E','D')
graph.add_edge('F','G')
nx.draw_networkx(graph)
```

## Directed Graph Network

```python
D=nx.DiGraph()
D.add_edge('A','B')
D.add_edge('B','C')
D.add_edge('C','F')
D.add_edge('C','E')
D.add_edge('E','D')
D.add_edge('F','G')
nx.draw_networkx(D)
```

## Weighted Graph Network

```python
G=nx.Graph()
G.add_edge('A','B', weight = 6)
G.add_edge('B','C', weight = 10)
G.add_edge('C','F', weight = 12)
G.add_edge('C','E', weight = 1)
G.add_edge('B','C', weight = 3)
G.add_edge('E','D', weight = 4)
G.add_edge('F','G', weight = 8)
nx.draw_networkx(G)
```

## Signed Network

```python
S=nx.Graph()
S.add_edge('A','B', sign = '+')
S.add_edge('B','C', sign = '+')
S.add_edge('C','F', sign = '-')
S.add_edge('C','E', sign = '+')
S.add_edge('B','C', sign = '+')
S.add_edge('E','D', sign = '-')
S.add_edge('F','G', sign = '-')
nx.draw_networkx(S)
```

## Relation Network

```python
R=nx.Graph()
R.add_edge('A','B', relation = 'mother')
R.add_edge('B','C', relation = 'friend')
R.add_edge('C','F', relation = 'father')
R.add_edge('C','E', relation = 'coworker')
R.add_edge('B','C', relation = 'friend')
R.add_edge('E','D', relation = 'sister')
R.add_edge('F','G', relation = 'friend')
```

## MultiGraph Network

```python
MG=nx.MultiGraph()
MG.add_edge('A','B', relation = 'mother')
MG.add_edge('A','B', relation = 'family')
MG.add_edge('B','C', relation = 'friend')
MG.add_edge('C','F', relation = 'father')
MG.add_edge('C','E', relation = 'coworker')
MG.add_edge('B','C', relation = 'friend')
MG.add_edge('E','D', relation = 'sister')
MG.add_edge('F','G', relation = 'friend')
```

## Graph with Weight and Attribute

```python
G = nx.Graph()
G.add_edge('A','B', weight=6, relation = 'Family')
G.add_edge('B','C', weight=13, relation = 'Friend')
G.add_edge('C','F', weight=21, relation = 'Coworker')
G.add_edge('C','E', weight=25, relation = 'Neighbor')
G.add_edge('F','G', weight=9, relation = 'Friend')
G.add_edge('E','D', weight=2, relation = 'Coworker')
G.add_edge('E','I', weight=10, relation = 'Coworker')
G.add_edge('E','J', weight=15, relation = 'Family')
G.add_edge('E','H', weight=9, relation = 'Coworker')
```

## List of all Edges

```python
G.edges()
```
