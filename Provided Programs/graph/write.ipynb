## Write graph object into Python pickle format

```python
# create a random graph
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
G = nx.path_graph(5)
nx.draw(G,with_labels=True)
```

```python
#write file to disk
nx.write_gpickle(G,'path.gpickle')
```

```python
import os
os.listdir()
```
