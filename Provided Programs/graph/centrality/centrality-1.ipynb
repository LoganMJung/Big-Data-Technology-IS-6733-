```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
k = nx.karate_club_graph()
nx.draw(k,node_color='yellow',node_size=100,with_labels=True)
```

```python
print(nx.info(k))
```

```python
dc=nx.degree_centrality(k)
print (dc)
print ("degree centrality =", max(zip(dc.values(), dc.keys())))
```

```python
bc=nx.betweenness_centrality(k)
print (bc)
print ("betweeness centrality =", max(zip(bc.values(), bc.keys())))
```

```python
ec=nx.betweenness_centrality(k)
print (ec)
print ("eigenvalue centrality =", max(zip(ec.values(), ec.keys())))
```

```python
cc=nx.closeness_centrality(k)
print (cc)
print ("closeness centrality =", max(zip(cc.values(), cc.keys())))
```
