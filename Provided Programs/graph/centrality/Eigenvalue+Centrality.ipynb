```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
G1 = nx.Graph()
G1.add_edges_from([(1, 6),
                   (1, 9),
                   (7, 0),
                   (4, 5),
                   (0, 2),
                   (2, 10),
                   (2, 14),
                   (2, 4),
                   (2, 3),
                   (4, 11),
                   (1,7),
                   (7, 8),
                   (7, 12),
                   (7, 13)])
nx.draw_networkx(G1,node_color='red', node_size=1000)
```

```python
centrality = nx.eigenvector_centrality(G1)
print (centrality)
```

```python
print ("centrality =", max(zip(centrality.values(), centrality.keys())))
```
