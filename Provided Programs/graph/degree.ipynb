```python
from networkx import *
import sys
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
n = 10 #nodes
m = 20 # edges

G = nx.gnm_random_graph(n,m)

print('Nodes Degrees')

for v in nodes(G):
    print('%s  %d  '%(v, degree(G,v)))
```
